# Single-Frequency HWPT Analysis

This notebook analyzes a single-frequency touchstone file (one point) and computes the maximum efficiency for that point.

The multi-frequency workflow lives in `sweep_case.ipynb`; this notebook is for the single-file case only.

```python
import os
import sys
from pathlib import Path

# Prefer local workspace package during development.
repo_root = Path.cwd().resolve().parent
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

if "COLAB_GPU" in os.environ:
    !pip install git+https://github.com/t-sasatani/wpt-tools
```

```python
from wpt_tools.data_classes import RichNetwork
from wpt_tools.analysis import nw_tools
```

### Load the single-frequency touchstone file

For a single-frequency network, `target_f` and `range_f` are optional: the only available point is used automatically.

```python
rich_nw = RichNetwork.from_touchstone("./assets/sample_single_freq.s2p")

f0 = float(rich_nw.nw.frequency.f[0])
print(f"frequency: {f0:.6e} Hz")
```

### Efficiency analysis at the single point

Frequency-domain plotting is skipped automatically (there is nothing to sweep), so the result is reported as a table.

```python
results = nw_tools.analyze_efficiency(
    rich_nw=rich_nw,
    rx_port=2,
    show_plot=True,
    show_data=True,
)

gmax = float(results.max_eff_opt)
print(f"\ngmax (maximum efficiency): {gmax:.12f}")
```
